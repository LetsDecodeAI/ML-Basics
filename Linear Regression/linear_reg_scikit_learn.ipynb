### 2. The Scikit-Learn Implementation Code

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression # The industry standard library

# 1. Generate Synthetic Data
# We create a simple linear relationship with some random noise
np.random.seed(42)
X = 4 * np.random.rand(100, 1) # Hours studied (0 to 4 hours)
y = 4 + 3 * X + np.random.randn(100, 1) # Exam score (y = 4 + 3x + noise)

# Visualize the data first
plt.scatter(X, y)
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.title("Student Data")
plt.show()
```

```python
# --- 2. The Scikit-Learn Magic ---

# Step A: Initialize the model
# This creates the mathematical container for our algorithm
model = LinearRegression()

# Step B: Train the model
# Remember our 'gradient_descent' function? 
# Scikit-learn does all that math inside this ONE line.
model.fit(X, y)

# Step C: Retrieve the parameters
# 'coef_' is our Weight (w) and 'intercept_' is our Bias (b)
final_w = model.coef_[0][0]
final_b = model.intercept_[0]

print(f"Learned Weight (w): {final_w:.2f}")
print(f"Learned Bias (b): {final_b:.2f}")
```

```python
# --- 3. Making Predictions ---
# Let's predict for a new student who studies 5 hours
new_X = np.array([[5]]) # Scikit-learn expects a 2D array
predicted_score = model.predict(new_X)

print(f"Prediction for 5 hours: {predicted_score[0][0]:.2f}")
```

```python
# --- 4. Visualization ---
# Use the model to predict for ALL data points to draw the line
line_y = model.predict(X)

plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, line_y, color='red', linewidth=3, label='Scikit-Learn Line')
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.legend()
plt.title("Linear Regression using Scikit-Learn")
plt.show()
```
